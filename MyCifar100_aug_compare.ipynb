```





from google.colab import drive
import pandas as pd

drive.mount('/content/gdrive')
#Then you can simply write to google drive as you would to a local file system like so:



hist_drp00 = pd.read_csv("/content/gdrive/My Drive/Colab Notebooks/MyCifar100/myresnet_cf100_model_1.csv")

hist_drp10 = pd.read_csv("/content/gdrive/My Drive/Colab Notebooks/MyCifar100/myresnet_cf100_model_1_aug.csv")

hist_drp20 = pd.read_csv("/content/gdrive/My Drive/Colab Notebooks/MyCifar100/myresnet_cf100_model_2_aug.csv")

hist_drp30 = pd.read_csv("/content/gdrive/My Drive/Colab Notebooks/MyCifar100/myresnet_cf100_model_3a_aug.csv")

```

```
import matplotlib.pyplot as plt
import numpy as np

#crn50 = crn50_iter50

X1 = hist_drp00
X2 = hist_drp10
X3 = hist_drp20
X4 = hist_drp30

colors = ['pink', 'purple','orange','red','blue','green','cyan','yellow']
index=0

plt.figure(0)
plt.plot(X1['val_acc'],colors[index])
plt.plot(X2['val_acc'],colors[index+1])
plt.plot(X3['val_acc'],colors[index+2])
plt.plot(X4['val_acc'],colors[index+3])

#plt.plot(X4['val_acc'],'purple')
#plt.plot(X5['val_acc'],'o')
#plt.plot(X6['val_acc'],'p')

labels = ['No Data Augmentation','With Data Augmentation','With Data Augmentation-2','With Data Augmentation-3']

title1 = "ResNet50+Dense,categorical_crossentropy \n w/wo data augmentation"
#plt.plot(crn50.history['acc'],'r')
#plt.plot(crn50.history['val_acc'],'g')

plt.xticks(np.arange(0, 100, 10.0))
plt.rcParams['figure.figsize'] = (8, 6)
plt.xlabel("Num of Epochs")
plt.ylabel("Validation Accuracy")

plt.title(title1+' Validation Accuracy')
plt.legend(labels)
 
plt.figure(1)
plt.plot(X1['val_loss'],colors[index])
plt.plot(X2['val_loss'],colors[index+1])
plt.plot(X3['val_loss'],colors[index+2])
plt.plot(X4['val_loss'],colors[index+3])
plt.xticks(np.arange(0, 100, 10.0))
plt.rcParams['figure.figsize'] = (8, 6)
plt.xlabel("Num of Epochs")
plt.ylabel("Validation Loss")
plt.title(title1+' valiation loss')

plt.legend(labels)
  
plt.show()
```
